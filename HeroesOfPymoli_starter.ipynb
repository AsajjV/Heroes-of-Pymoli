## Purchasing Analysis (Gender)

```python
#in order to get the correct numbers this should only include unique SNs, not the entire list
heroes_df

#remove duplicate heroes
heroes_no_dups_df = heroes_df.drop_duplicates("SN")

heroes_no_dups_df

gender_grouped_df = heroes_no_dups_df.groupby(["Gender"])

gender_grouped_df.count()

gender_count_df = pd.DataFrame(gender_grouped_df["SN"].count())
gender_count_df=gender_count_df.rename(columns=({"SN":"Total"}))
gender_count_df

gender_percent_df=pd.DataFrame(gender_count_df["Total"]/sum(gender_count_df["Total"])*100)

gender_percent_df=gender_percent_df.rename(columns=({"Total":"Percentage"}))
gender_percent_df

merged_df=pd.merge(gender_count_df, gender_percent_df, on="Gender")
merged_df

merged_df["Percentage"]=merged_df["Percentage"].map("{:.2f}".format)
merged_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
heroes_grpby_gender_df=heroes_df.groupby(["Gender"])
heroes_grpby_gender_df.head()

print(heroes_grpby_gender_df)
purchase_count_by_gender = heroes_grpby_gender_df.count()
purchase_count_by_gender_df = pd.DataFrame(purchase_count_by_gender["Purchase ID"])
purchase_count_by_gender_df

avg_purchase_by_gender_df = pd.DataFrame(heroes_grpby_gender_df["Price"].mean())
avg_purchase_by_gender_df

total_purchase_by_gender = heroes_grpby_gender_df["Price"].sum()
total_purchase_by_gender_df = pd.DataFrame(heroes_grpby_gender_df["Price"].sum())
total_purchase_by_gender_df

gender_count = gender_grouped_df["SN"].count()
gender_count

avg_total_purch_by_gender_df = pd.DataFrame(total_purchase_by_gender/gender_count)
avg_total_purch_by_gender_df 

summary_df = pd.merge(total_purchase_by_gender_df, avg_total_purch_by_gender_df, on="Gender")
summary_df = summary_df.rename(columns=({0:"Avg Total Purchase per Person"}))
summary_df = summary_df.rename(columns=({"Price":"Total Purchase Value"}))

summary_df = pd.merge(avg_purchase_by_gender_df, summary_df, on="Gender")
summary_df = summary_df.rename(columns=({"Price":"Average Purchase Price"}))
summary_df = pd.merge(purchase_count_by_gender_df, summary_df, on="Gender")
summary_df = summary_df.rename(columns=({"Purchase ID":"Purchase Count"}))

summary_df["Average Purchase Price"]=summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Avg Total Purchase per Person"]=summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)
summary_df["Total Purchase Value"]=summary_df["Total Purchase Value"].map("${:.2f}".format)

summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Age_df = pd.DataFrame(heroes_no_dups_df)

bins = [0, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5, 39.5, 100]
group_names = ["< 10", 
               "10-14", 
               "15-19", 
               "20-24", 
               "25-29", 
               "30-34", 
               "35-39", 
               "40+"]
Age_df = pd.DataFrame(pd.cut(Age_df["Age"], bins, labels=group_names))
Age_df.head()


Age_group = Age_df.groupby(["Age"])
#print(Age_group)
print(Age_group["Age"].count())

#create a df of the bins
age_group_df = pd.DataFrame(Age_group["Age"].count())
age_group_df = age_group_df.rename(columns={ age_group_df.columns[0]: "Count" })

age_group_df

age_grp = (Age_group["Age"].count())
age_grp

#create a df of percents
age_grp_p_df = pd.DataFrame((age_grp/sum(age_grp))*100)
age_grp_p_df = age_grp_p_df.rename(columns={ age_grp_p_df.columns[0]: "Percent" })
age_grp_p_df

age_summary_df = pd.merge(age_group_df, age_grp_p_df, on=["Age"])
#age_summary_df = age_summary_df.rename(columns = {"SN_x":"Total Count", "SN_y": "Percent of Players"})

#format percent
age_summary_df["Percent"]=age_summary_df["Percent"].map("{:.2f}".format)

age_summary_df
```

```python
Age1_df = pd.DataFrame(heroes_df)

bins = [0, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5, 39.5, 100]
group_names = ["<10", 
               "10-14", 
               "15-19", 
               "20-24", 
               "25-29", 
               "30-34", 
               "35-39", 
               "40+"]

Age1_df["Age Ranges"] = pd.cut(Age1_df["Age"], bins, labels=group_names)
Age1_df.head()

heroes_grpby_age_df=Age1_df.groupby(["Age Ranges"])
heroes_grpby_age_df.head()

print(heroes_grpby_age_df)
purchase_count_by_age = heroes_grpby_age_df.count()
purchase_count_by_age_df = pd.DataFrame(purchase_count_by_age["Purchase ID"])
purchase_count_by_age_df

#Average purchase price
avg_purchase_by_age_df = pd.DataFrame(heroes_grpby_age_df["Price"].mean())
avg_purchase_by_age_df

total_purchase_by_age = heroes_grpby_age_df["Price"].sum()
total_purchase_by_age_df = pd.DataFrame(heroes_grpby_age_df["Price"].sum())
total_purchase_by_age

#age_grp
#total_purchase_by_age/age_grp

avg_total_purch_by_age_df = pd.DataFrame(total_purchase_by_age/age_grp)
avg_total_purch_by_age_df.index.name = "Age Ranges"
avg_total_purch_by_age_df

##Note to whoever happens to have the joy of grading this - I cannot figure out why I am getting 2
##"<10" age ranges in this df, or why they come up as NaN.  I put in my best effort.

age1_summary_df = pd.merge(purchase_count_by_age_df, avg_purchase_by_age_df, on = "Age Ranges")
age1_summary_df = age1_summary_df.rename(columns=({"Purchase ID":"Purchase Count", "Price":"Average Purchase Price"}))
age1_summary_df = pd.merge(age1_summary_df, total_purchase_by_age_df, on="Age Ranges")
age1_summary_df = age1_summary_df.rename(columns=({"Price":"Total Purchase Value"}))
age1_summary_df = pd.merge(age1_summary_df, avg_total_purch_by_age_df, on="Age Ranges")
age1_summary_df = age1_summary_df.rename(columns=({0:"Avg Total Purchase by Person"}))
age1_summary_df

#format the df
age1_summary_df["Average Purchase Price"]=age1_summary_df["Average Purchase Price"].map("${:.2f}".format)
age1_summary_df["Avg Total Purchase by Person"]=age1_summary_df["Avg Total Purchase by Person"].map("${:.2f}".format)
age1_summary_df["Total Purchase Value"]=age1_summary_df["Total Purchase Value"].map("${:.2f}".format)

age1_summary_df

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
hop_df.columns

top_seller_df = hop_df[["Item ID", "Item Name", "Price","Purchase ID"]]
top_seller_df.head()

Item_group_df = hop_df.groupby(["Item ID", "Item Name", "Price"])
print(Item_group_df)
#Item_group_df.count
print(Item_group_df.count)

Item_group_df.count()

#Purchase Count
purchase_count_by_item = Item_group_df.count()
purchase_count_by_item
purchase_count_by_item_df = pd.DataFrame(purchase_count_by_item["Purchase ID"])
purchase_count_by_item_df.head()
purchase_count_by_item_df = purchase_count_by_item_df.rename(columns = ({"Purchase ID":"Purchase Count"}))
#purchase_count_by_item_df = purchase_count_by_item_df.columns({"Item ID", "Item Name", "Purchase Count", "Price"})
purchase_count_by_item_df.head(5)



#total purchase value

#print(Item_group_df.sum())
print(Item_group_df["Price"].sum)

total_purchase_value_df = pd.DataFrame(Item_group_df["Price"].sum())
total_purchase_value_df= total_purchase_value_df.rename(columns = ({"Price":"Total Purchase Value"}))
total_purchase_value_df.head(5)


most_pop_summary_df = pd.merge(purchase_count_by_item_df, total_purchase_value_df, how = "left", 
                               left_on =["Item ID", "Item Name", "Price"], 
                              right_on = ["Item ID", "Item Name", "Price"])

most_pop_summary_df

most_pop_summary_df = most_pop_summary_df.reset_index(["Item ID", "Item Name", "Price"])
most_pop_summary_df.columns
most_pop_summary_df = most_pop_summary_df[["Item ID", "Item Name", "Purchase Count", "Price", "Total Purchase Value"]]
most_pop_summary_df = most_pop_summary_df.set_index(["Item ID"])

most_pop_summary_df["Price"]=most_pop_summary_df["Price"].map("${:.2f}".format)
##Removing formatting for next section sort
#most_pop_summary_df["Total Purchase Value"]=most_pop_summary_df["Total Purchase Value"].map("${:.2f}".format)

most_pop_summary_df.nlargest(5,"Purchase Count")
```
